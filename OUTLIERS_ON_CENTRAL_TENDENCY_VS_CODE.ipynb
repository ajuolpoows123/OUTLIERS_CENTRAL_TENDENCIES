```python
import numpy as np
import docx

# Specify the file path
file_path = r"C:\Users\jamar\Downloads\cholesterol_dataset_DSC__510.docx"

# Load the document and extract data from tables
cholesterol_levels = []
doc = docx.Document(file_path)

# Iterate through each table in the document
for table in doc.tables:
    for row in table.rows:
        for cell in row.cells:
            line = cell.text.strip()
            try:
                # Try to convert cell content to float
                cholesterol_levels.append(float(line))
            except ValueError:
                # Ignore lines that can't be converted to float
                continue

# Verify loaded data
print("Cholesterol Levels Loaded:", cholesterol_levels)

# Calculate mean and median
if cholesterol_levels:
    mean = np.mean(cholesterol_levels)
    median = np.median(cholesterol_levels)
    print(f"Mean: {mean}, Median: {median}")
else:
    print("No valid cholesterol data found.")
```

```python
import numpy as np
import pandas as pd
import docx
import matplotlib.pyplot as plt

# Specify the file path
file_path = r"C:\Users\jamar\Downloads\cholesterol_dataset_DSC__510.docx"

# Load the document and extract data from tables
cholesterol_levels = []
doc = docx.Document(file_path)

# Iterate through each table in the document
for table in doc.tables:
    for row in table.rows:
        for cell in row.cells:
            line = cell.text.strip()
            try:
                # Try to convert cell content to float
                cholesterol_levels.append(float(line))
            except ValueError:
                # Ignore lines that can't be converted to float
                continue

# Verify loaded data
print("Cholesterol Levels Loaded:", cholesterol_levels)

# Calculate mean and median
if cholesterol_levels:
    mean = np.mean(cholesterol_levels)
    median = np.median(cholesterol_levels)
    print(f"Mean: {mean}, Median: {median}")
else:
    print("No valid cholesterol data found.")

# Plotting
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(cholesterol_levels, bins=10, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=1, label=f'Median: {median:.2f}')
plt.title('Histogram of Cholesterol Levels')
plt.xlabel('Cholesterol Level')
plt.ylabel('Frequency')
plt.legend()
```
